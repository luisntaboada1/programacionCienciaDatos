***Ventajas de usar VSCode como interfaz para git y ambientes virtuales***

***Ventajas de usar VSCode***

1. Multilenguaje en un solo lugar:
   VSCode permite correr codigo de diferentes lenguajes de programacion como Python, C/C++, Rust, entre muchos
2. Terminal integrada:
   Cuando corres, tienes la terminal integrada para poder detectar errores o ver los outputs de tu programa
3. Depurar:
   VSCode ofrece herramientas de debigging para poder analizar procesos y detectar errores o fallas en el codigo de manera detallada
4. Extensiones:
   Yo creo que esta es la mas importante. Al ser abierto, VSCode ofrece una cantidad inimaginable de extensiones que te permiten configurar el      ambiente de desarrollo a tu medida.

Algunas de las ventajas de controlar repositorios git y github a traves de VSCode son:
1. Principalmente es la simplificacion de acciones con botones:
   Al usar VSCode para controlar repositorios github, en vez de escribir comandos para cada accion que se quiera realizar, basta con presionar      un boton
2. Commits y ramas con un clic:
   Interfaz visual para stage, commit, revert y manejo de ramas sin comandos.
3. Resolucion de conflictos:
   editor lateral con resaltado y botones “Accept Current/Incoming” muy claro.
entre mas funcionalidades

***Seaborn***

```python
import seaborn as sns
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
```
