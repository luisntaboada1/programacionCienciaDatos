***PANDAS Y MATPLOTLIB***

Voy a usar un dataset que arme yo personalmente. Es un resumen de las ultimas 18 temporadas de formula 1. Recopile los pilotos, las posiciones obtenidas en cada carrera, el numero de victorias, podios, puntos conseguidos, y mas de cada equipo que ha competido desde el 2007 hasta la temporada pasada (2024). Este dataset se armo usando los datos de un dataset mucho mas grande.

```python
#Pandas
import pandas as pd
import matplotlib.pyplot as plt

constructors = pd.read_csv('../../../datasets/constructors.csv')
all_seasons = pd.read_csv('../../../datasets/constructor_season_summary.csv')

season = all_seasons[all_seasons['year'] == 2021]
season_teams = pd.merge(constructors, season, on='constructorId')

team_names = season_teams['constructorRef'].to_list()
team_points = season_teams['final_constructor_points'].to_list()

print(team_names)
print(team_points)
```

```python
# Matplotlib

fig, ax = plt.subplots()
ax.bar(team_names, team_points, width=1, edgecolor='white', linewidth=2)
ax.set_ylabel('Points')
ax.set_xlabel('Team')
ax.set_title('Constructor Standings')
plt.xticks(rotation=90, ha='right')
```
